```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
w, h = 8, 6
def unit_step(a, n):
    unit =[]
    for sample in n:
        if sample<a:
            unit.append(0)
        else:
            unit.append(1)
    return(unit)
  
a = 0
UL = 5
LL = -5
n = np.arange(LL, UL, 1)
unit = unit_step(a, n)
fig, axe = plt.subplots(2, 2, figsize=(16, 8))
fig.tight_layout(pad=5)
fig = plt.subplot(2,2, 1)
plt.stem(n, unit)
plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.title('Unit step sequence')
plt.grid()


def unit_impulse(a, n):
    delta =[]
    for sample in n:
        if sample == a:
            delta.append(1)
        else:
            delta.append(0)
              
    return delta
  
a = 0
UL = 5
LL = -5
n = np.arange(LL, UL, 1)
d = unit_impulse(a, n)
fig = plt.subplot(2,2, 2)
plt.stem(n, d)

plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.title('Unit Sequence')
plt.grid()


def exponential(a, n):
    expo =[]
    for sample in n:
        expo.append(np.exp(a * sample))
    return (expo)
         
a = -0.5
UL = 5
LL = -5
n = np.arange(LL, UL, 1)
x = exponential(a, n)
fig = plt.subplot(2,2, 3)
plt.stem(n, x)

plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.title('Exponential Sequence')
plt.grid()


def sine(n):
    expo =[]
    for sample in n:
        expo.append(2*np.sin(np.pi * sample/5))
    return (expo)
         
a = -1
UL = 10
LL = -10
n = np.arange(LL, UL, 1)
x = sine(n)
fig = plt.subplot(2,2, 4)
plt.stem(n, x)

plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.title('Sinusoidal Sequence')
plt.grid()

plt.savefig('bai1.png')
plt.show()
```
